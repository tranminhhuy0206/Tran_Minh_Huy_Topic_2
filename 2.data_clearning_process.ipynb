3. CLEARNING DATA PROCESS


3.1 Standardization & Handle Missing Value

```python
## 3.1 Standardization & Handle Missing Value

# 1. Ensure Income is a numeric type and NaN is handled (If Income can be a string/object)
df_marketing_campaign['Income'] = pd.to_numeric(
    df_marketing_campaign['Income'],
    errors='coerce' # Replace non-numeric values ​​with NaN
)

# 2. Replace NaN in Income with mean
df_marketing_campaign['Income'].fillna(df_marketing_campaign['Income'].mean(), inplace=True)

# 3. Remove duplicate rows
df_marketing_campaign.drop_duplicates(inplace=True)

# 4. Remove missing rows in other columns
df_marketing_campaign.dropna(inplace=True)

print(f"Kích thước DataFrame sau khi xử lý NaN và Trùng lặp: {df_marketing_campaign.shape}")
df_marketing_campaign.info()
print(df_marketing_campaign.columns)
```

3.3 Standardize capitalization

```python
column_mapping = {
    'ID': 'customer_id',              
    'Year_Birth': 'year_birth',
    'Education': 'education',
    'Marital_Status': 'marital_status',
    'Income': 'income',
    'Kidhome': 'kidhome',
    'Teenhome': 'teenhome',
    'Dt_Customer': 'dt_customer',
    'Recency': 'recency',
    
    # Amount
    'MntWines': 'amount_wines',
    'MntFruits': 'amount_fruits',
    'MntMeatProducts': 'amount_meat_products',
    'MntFishProducts': 'amount_fish_products',
    'MntSweetProducts': 'amount_sweet_products',
    'MntGoldProds': 'amount_gold_prods',
    
    # Number
    'NumDealsPurchases': 'num_deals_purchases',
    'NumWebPurchases': 'num_web_purchases',
    'NumCatalogPurchases': 'num_catalog_purchases',
    'NumStorePurchases': 'num_store_purchases',
    'NumWebVisitsMonth': 'num_web_visits_month',
    
    # Campaign
    'AcceptedCmp3': 'accepted_campaign_3',
    'AcceptedCmp4': 'accepted_campaign_4',
    'AcceptedCmp5': 'accepted_campaign_5',
    'AcceptedCmp1': 'accepted_campaign_1',
    'AcceptedCmp2': 'accepted_campaign_2',
    
    'Complain': 'complain',
    'Z_CostContact': 'z_cost_contact',
    'Z_Revenue': 'z_revenue',
    'Response': 'response'
}

# Apply rename
df_marketing_campaign.rename(columns=column_mapping, inplace=True)

print(df_marketing_campaign.columns)
```

3.4 Convert data type

```python
# Convert dt_customer to datetime
df_marketing_campaign['dt_customer'] = pd.to_datetime(df_marketing_campaign['dt_customer'], format='%d-%m-%Y', errors='coerce')

# Calculate age
current_year = datetime.now().year
df_marketing_campaign['age'] = current_year - df_marketing_campaign['year_birth']

# Customer_tenure
df_marketing_campaign['customer_tenure_days'] = (datetime.now() - df_marketing_campaign['dt_customer']).dt.days

# Biz Sense
df_marketing_campaign = df_marketing_campaign[df_marketing_campaign['age'] < 100]

print(df_marketing_campaign[['year_birth', 'dt_customer', 'age', 'customer_tenure_days']].head())
```

3.5 Detect & Remove Outliers Using IQR

```python
# Make sure Income is numeric and does not have NaN values
df_marketing_campaign['income'] = pd.to_numeric(
    df_marketing_campaign['income'], 
    errors='coerce' 
).fillna(df_marketing_campaign['income'].mean()) 

# 1. Calculate Q1, Q3, and IQR
Q1 = df_marketing_campaign['income'].quantile(0.25)
Q3 = df_marketing_campaign['income'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 2. Identify outliers
outliers_iqr = df_marketing_campaign[(df_marketing_campaign['income'] < lower_bound) | (df_marketing_campaign['income'] > upper_bound)]
print(f"Outliers using IQR method (income): {outliers_iqr.shape[0]} rows found.")
# print(outliers_iqr[['ID', 'Income']]) 

# 3. Remove outliers
df_marketing_campaign = df_marketing_campaign[(df_marketing_campaign['income'] >= lower_bound) & (df_marketing_campaign['income'] <= upper_bound)]
print("\nDataFrame after removing outliers (IQR method):")
print(df_marketing_campaign.shape)
```

4. Save Clear Data Into Database To Visiualization ti

```python
import pyodbc

# --- Connect ---
conn = pyodbc.connect(
        "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=DESKTOP-MV8RPS5;"
    "DATABASE=Data_Topic;"
    "Trusted_Connection=yes;"
    "TrustServerCertificate=yes;"
)
cursor = conn.cursor()

# --- Map pandas dtype -> SQL Server type ---
dtype_map = lambda dt: "INT" if pd.api.types.is_integer_dtype(dt) else \
                       "FLOAT" if pd.api.types.is_float_dtype(dt) else \
                       "BIT" if pd.api.types.is_bool_dtype(dt) else \
                       "DATETIME" if pd.api.types.is_datetime64_any_dtype(dt) else \
                       "VARCHAR(255)"

table_name = "data_clean_maketing_campaign"

# --- Drop + Create table ---
cursor.execute(f"IF OBJECT_ID('dbo.{table_name}', 'U') IS NOT NULL DROP TABLE dbo.{table_name}")
cols = ", ".join([f"[{c}] {dtype_map(df_marketing_campaign[c].dtype)}" for c in df_marketing_campaign.columns])
cursor.execute(f"CREATE TABLE dbo.{table_name} ({cols})")

# --- Insert rows ---
cursor.fast_executemany = True
cursor.executemany(f"INSERT INTO dbo.{table_name} VALUES ({','.join(['?']*len(df_marketing_campaign.columns))})", df_marketing_campaign.values.tolist())
conn.commit()
```
